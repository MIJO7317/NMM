<b>Задание:</b><br>
Найти аналитическое решение (методом Фурье) и численное решение по неявной схеме задач:
\begin{equation*}
\begin{cases}
u_t = a^2 u_{xx}, \quad 0 < x < l, \quad t > 0,\\
u(x,0) = u_0(x),\\
u(0,t) = u_1(t),\\
u(l,t) = u_2(t).
\end{cases}
\end{equation*}

<b>Аналитическое решение:</b><br>

Сделаем замену $u(x,t) = v(x,t) + b(t)x+c(t)$, где $b(t)$ и $c(t)$ - функции, которые мы подберем позже. Тогда
\begin{equation*}
\begin{cases}
v_t + b(t) = a^2 v_{xx}, \quad 0 < x < l, \quad t > 0,\\
v(x,0) + b(0)x+c(0) = u_0(x),\\
v(0,t) +c(t) = u_1(t),\\
v(l,t) + b(t)l+c(t) = u_2(t).
\end{cases}
\end{equation*}
Подберем $b(t)$ и $c(t)$ так, чтобы граничные условия были равны нулю:
\begin{equation*}
\begin{cases}
b(t) = \frac{u_2(t)-u_1(t)}{l},\\
c(t) = u_1(t).
\end{cases}
\end{equation*}
Тогда
\begin{equation*}
\begin{cases}
v_t = a^2 v_{xx} + f(t),\quad f(t) = - \frac{u_2(t)-u_1(t)}{l}, \quad 0 < x < l, \quad t > 0,\\
v(x,0) = u_0(x) - \frac{u_2(0)-u_1(0)}{l}x-u_1(0),\\
v(0,t) = 0,\\
v(l,t) = 0.
\end{cases}
\end{equation*}

После разделения переменных $v(x,t) = X(x)T(t)$ получим, что задача Штурма-Лиувилля для функции $X(x)$ имеет вид
\begin{equation*}
\begin{cases}
X''(x) + \lambda X(x) = 0, \quad 0 < x < l,\\
X(0) = X(l) = 0.
\end{cases}
\end{equation*}

Общее решение этой задачи имеет вид
\begin{equation*}
X(x) = C_1 \sin \sqrt{\lambda}x + C_2 \cos \sqrt{\lambda}x.
\end{equation*}

Подставим это решение в граничные условия:
\begin{equation*}
\begin{cases}
C_2 = 0,\\
C_1 \sin \sqrt{\lambda}l = 0.
\end{cases}
\end{equation*}

Отсюда получаем, что $\sqrt{\lambda} = \frac{\pi n}{l}$, где $n = 1,2,3,...$.

Собственные значения $\lambda_n = \left(\frac{\pi n}{l}\right)^2, \quad n = 1,2,3,...$.

Тогда собственные функции имеют вид
\begin{equation*}

\end{equation*}

\begin{equation*}
v(x,t) = \sum_{n=1}^{\infty} T_n(t) X_n(x).
\end{equation*}

Также представим $f(t)$ в виде ряда Фурье:
\begin{equation*}
f(t) = \sum_{n=1}^{\infty} f_n(t) X_n(x), \quad f_n(t) = \frac{2}{l} \int_0^l f(t) X_n(x) dx.
\end{equation*}

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Определение функций и параметров
l = 1
T = 1
u_0 = lambda x: np.sin(np.pi * x)

# Граничные условия 3 рода
# alpha_1 * u_x(0, t) + beta_1 * u(0, t) = gamma_1(t)
# alpha_2 * u_x(l, t) + beta_2 * u(l, t) = gamma_2(t)
alpha_1 = 0
alpha_2 = 0
beta_1 = 1
beta_2 = 1
gamma_1 = lambda t: 0
gamma_2 = lambda t: 0

f = lambda x, t: 0
k = lambda u, x, t: 1

# Задание сетки
N = 101
M = 101
xs = np.linspace(0, l, N)
ts = np.linspace(0, T, M)
h = xs[1] - xs[0]
tau = ts[1] - ts[0]

# Задание начального условия
u = np.zeros((M, N))
u[0] = u_0(xs)

# Цикл по времени
for m in range(M - 1):
    # Определение матрицы A и вектора b
    A = np.zeros((N, N))
    b = np.zeros(N)
    A[0][0] = beta_1 - 3/2 * alpha_1 / h
    A[0][1] = 2 * alpha_1 / h
    A[0][2] = -alpha_1 / (2 * h)
    b[0] = gamma_1(ts[m])
    A[N - 1][N - 3] = alpha_2 / (2 * h)
    A[N - 1][N - 2] = -2 * alpha_2 / h
    A[N - 1][N - 1] = beta_2 + 3/2 * alpha_2 / h
    b[N - 1] = gamma_2(ts[m])
    for n in range(1, N - 1):
        A[n][n - 1] = -1 / h ** 2 * (k(u[m][n], xs[n], ts[m]+tau/2) + k(u[m][n-1], xs[n-1], ts[m]+tau/2)) / 2
        A[n][n] = 2 / h ** 2 * k(u[m][n], xs[n], ts[m]+tau/2) + 1 / tau
        A[n][n + 1] = -1 / h ** 2 * (k(u[m][n], xs[n], ts[m]+tau/2) + k(u[m][n+1], xs[n+1], ts[m]+tau/2)) / 2
        b[n] = u[m][n] / tau + f(xs[n], ts[m])
    # Решение системы
    u[m + 1] = np.linalg.solve(A, b)

# Построение графика
plt.figure(figsize=(12, 5))
for m in range(M):
    if m % 10 == 0:
        plt.plot(xs, u[m], label='t = ' + str(int(100*ts[m])/100))
plt.legend()
plt.grid()
plt.xlabel('x')
plt.ylabel('u')
plt.show()
```

```python
u_an = lambda x, t: np.exp(-np.pi**2 * t) * np.sin(np.pi * x)
# Plot error between analytical and numerical solutions
plt.figure(figsize=(12, 5))
for m in range(M):
    if m % 10 == 0:
        plt.plot(xs, np.abs(u_an(xs, ts[m]) - u[m]), label='t = ' + str(int(100*ts[m])/100))
plt.legend()
plt.grid()
plt.xlabel('x')
plt.ylabel('Error')
plt.show()
```

```python
import numpy as np
import matplotlib
import matplotlib.pyplot as plt
matplotlib.rcParams["figure.dpi"] = 100
import plotly.graph_objects as go
from scipy.optimize import fsolve
from mpl_toolkits.mplot3d import Axes3D
from matplotlib import animation
```

```python
a = 0
b = 1
c = 0
d = 1
T = 2

k_x = lambda x, y : 1
k_y = lambda x, y : 1
q_x = lambda x, y : 1
q_y = lambda x, y : 1

f = lambda x, y, t : 0

mu_a = lambda y, t : 0
mu_b = lambda y, t : 0
mu_c = lambda x, t : 0
mu_d = lambda x, t : 0

u0 = lambda x, y : 1

tau = 0.01
h_x = 0.05
h_y = 0.05
M = int(T/tau) + 1
N_x = int((b-a)/h_x) + 1
N_y = int((d-c)/h_y) + 1
U = np.zeros((M, N_x, N_y))
Xs = np.linspace(a, b, N_x)
Ys = np.linspace(c, d, N_y)
Ts = np.linspace(0, T, M)

for i in range(N_x):
    for j in range(N_y):
        U[0][i][j] = u0(i*h_x, j*h_y)

for n in range(1, M):
    U_half = np.zeros((N_x, N_y))
    for j in range(0, N_y):
        if j == 0:
            U_half[:,0] = np.fromiter(map(lambda x : mu_c(x, Ts[n]), Xs), float)
        elif j == N_y - 1:
            U_half[:,N_y - 1] = np.fromiter(map(lambda x : mu_d(x, Ts[n]), Xs), float)
        else:
            A = np.zeros((N_x, N_x))
            B = np.zeros(N_x)
            for i in range(0, N_x):
                if i == 0:
                    A[0][0] = 1
                    B[0] = mu_a(Ys[j], Ts[n])
                elif i == N_x-1:
                    A[N_x - 1][N_x - 1] = 1
                    B[N_x - 1] = mu_b(Ys[j], Ts[n])
                else:
                    A[i][i-1] = tau/2/h_x**2*q_x(Xs[i], Ys[j])*k_x((Xs[i]+Xs[i-1])/2, Ys[j])
                    A[i][i] = -(1 + tau/2/h_x**2*q_x(Xs[i], Ys[j])*(k_x((Xs[i]+Xs[i+1])/2, Ys[j]) + k_x((Xs[i]+Xs[i-1])/2, Ys[j])))
                    A[i][i+1] = tau/2/h_x**2*q_x(Xs[i], Ys[j])*k_x((Xs[i]+Xs[i+1])/2, Ys[j])
                    B[i] = -(U[n-1][i][j] + tau/2 * (f(Xs[i], Ys[j], (Ts[n]+Ts[n-1])/2) + q_y(Xs[i], Ys[j])/h_y**2*(k_y(Xs[i], (Ys[j]+Ys[j+1])/2)*(U[n-1][i][j+1] - U[n-1][i][j]) - k_y(Xs[i], (Ys[j]+Ys[j-1])/2)*(U[n-1][i][j] - U[n-1][i][j-1]))))
            U_half[:,j] = np.linalg.solve(A, B)
    for i in range(0, N_x):
        if i == 0:
            U[n][0] = np.fromiter(map(lambda y : mu_a(y, Ts[n]), Ys), float)
        elif i == N_x - 1:
            U[n][N_x - 1] = np.fromiter(map(lambda y : mu_b(y, Ts[n]), Ys), float)
        else:
            A = np.zeros((N_y, N_y))
            B = np.zeros(N_y)
            for j in range(0, N_y):
                if j == 0:
                    A[0][0] = 1
                    B[0] = mu_c(Xs[i], Ts[n])
                elif j == N_y-1:
                    A[N_y - 1][N_y - 1] = 1
                    B[N_y - 1] = mu_d(Xs[i], Ts[n])
                else:
                    A[j][j-1] = tau/2/h_y**2*q_y(Xs[i], Ys[j])*k_y(Xs[i], (Ys[j]+Ys[j-1])/2)
                    A[j][j] = -(1 + tau/2/h_y**2*q_y(Xs[i], Ys[j])*(k_y(Xs[i], (Ys[j]+Ys[j-1])/2) + k_y(Xs[i], (Ys[j]+Ys[j+1])/2)))
                    A[j][j+1] = tau/2/h_y**2*q_y(Xs[i], Ys[j])*k_y(Xs[i], (Ys[j]+Ys[j+1])/2)
                    B[j] = -(U_half[i][j] + tau/2 * (f(Xs[i], Ys[j], Ts[n]) + q_x(Xs[i], Ys[j])/h_x**2*(k_x((Xs[i] + Xs[i+1])/2, Ys[j])*(U_half[i+1][j] - U_half[i][j]) - k_x((Xs[i]+Xs[i-1])/2, Ys[j])*(U_half[i][j] - U_half[i-1][j]))))
            U[n][i] = np.linalg.solve(A, B)
```
